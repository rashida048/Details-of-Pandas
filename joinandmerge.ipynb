```python
import numpy as np
def energy():
    energy = pd.read_excel(r'C:\Users\User\Desktop\New folder\Launchcode\coursera dataScience\course1 week3\Energy Indicators.xls')[16:243]
    energy = energy.drop(['Unnamed: 0','Unnamed: 1'], axis = 1)
    energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    energy['Country'] = energy['Country'].str.replace(r'\d+','')
    energy['Energy Supply'] = energy['Energy Supply'].apply(lambda x: np.NaN if x=='...' else x)
    energy['Energy Supply'] = energy['Energy Supply']*1000000
    dicts = {"Republic of Korea": "South Korea",
                   "United States of America": "United States",
                   "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
                   "China, Hong Kong Special Administrative Region": "Hong Kong"}
    energy['Country'] = energy['Country'].apply(lambda x: dicts[x] if x in dicts else x)
    energy['Country'] = energy['Country'].str.replace(r"\(.*\)","")
    return energy
energy = energy()
energy.head()
```

```python
def gdp():
    GDP = pd.read_csv('world_bank.csv', skiprows = 4)
    dicts = {"Korea, Rep.": "South Korea", 
           "Iran, Islamic Rep.": "Iran",
           "Hong Kong SAR, China": "Hong Kong"}
    GDP['Country Name'] = GDP['Country Name'].apply(lambda x: dicts[x] if x in dicts else x)
    colkeep = ['Country Name', '2006', '2007', '2008','2009','2010', '2011','2012','2013','2014','2015']
    GDP = GDP[colkeep]
    return GDP
GDP = gdp()
GDP.head()
```

```python
def scimen():
    ScimEn = pd.read_excel('scimagojr-3.xlsx')
    return ScimEn
ScimEn = scimen()
ScimEn.head()
```
